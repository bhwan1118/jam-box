```python
# These are the midi notes that correspond to all the white keys
note = [i for i in range(28,101)]

#Generating the midi for full range

white = []
for i in note:
    if i % 12 == 0:
        print(i,"C")
        white.append(i)
    if (i - 2) % 12 == 0:
        print(i, "D")
        white.append(i)
    if (i - 4) % 12 == 0:
        print(i, "E")
        white.append(i)
    if (i - 5) % 12 == 0:
        print(i, "F")
        white.append(i)

    if (i - 7) % 12 == 0:
        print(i, "G")
        white.append(i)

    if (i - 9) % 12 == 0:
        print(i, "A")
        white.append(i)

    if (i - 11) % 12 == 0:
        print(i, "B")  
        white.append(i)


```

```python
# Generate the note strings

Letters = ["C", "CS", "D", "DS", "E", "F", "FS", "G", "GS", "A", "AS", "B"]

gen_notes = []

for i in range(10):
    for j in range(12):
        gen_notes.append("NOTE_" + str(Letters[j]) + str(i))
```

```python
# General Script for generating all Major scales starting at C Maj : 0

#all major scales stored:
majors = []
scale = []

for j in range(12):
    for i in note:
        if i % 12 == 0:
            scale.append(gen_notes[i + j])
            #print(i, gen_notes[i + 2])
        if (i - 2) % 12 == 0:
            #print(i, gen_notes[i + 2])
            scale.append(gen_notes[i + j])
        if (i - 4) % 12 == 0:
            scale.append(gen_notes[i + j])    
            #print(i, gen_notes[i + 2])
        if (i - 5) % 12 == 0:
            scale.append(gen_notes[i + j])        
            #print(i, gen_notes[i + 2])
        if (i - 7) % 12 == 0:
            scale.append(gen_notes[i + j])
            #print(i, gen_notes[i + 2])
        if (i - 9) % 12 == 0:
            scale.append(gen_notes[i + j])
            #print(i, gen_notes[i + 2])
        if (i - 11) % 12 == 0:
            scale.append(gen_notes[i + j])
            #print(i, gen_notes[i + 2])    
    majors.append(scale)
    scale = []
```
